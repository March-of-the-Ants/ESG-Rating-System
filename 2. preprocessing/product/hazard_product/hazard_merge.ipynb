2. 데이터 불러오기

```python
hazard_2020 = pd.read_excel('./위해상품2020.xlsx')
hazard_2021 = pd.read_excel('./위해상품2021.xlsx')

kosdaq = pd.read_excel('../../ESG-Rating-System/crawling/dart/kosdaq_report_final.xlsx')
kosdaq_df = pd.DataFrame({'사업자번호':kosdaq['사업자등록번호'], '기업명':kosdaq['회사명']})
```

```python
kosdaq_df
# 사업자번호 : int에서 str으로 type 변경
kosdaq_df['사업자번호'] = kosdaq_df['사업자번호'].astype(str)
type(kosdaq_df['사업자번호'][8])
```

```python
hazard_2020['사업자번호'].replace('-', '', regex=True, inplace=True)
hazard_2020
```

```python
hazard_2020_df = pd.DataFrame({'사업자번호':hazard_2020['사업자번호'], '등록일시':hazard_2020['등록일시']})
hazard_2020_df
```

```python
hazard_2020_count = hazard_2020_df.groupby('사업자번호').count().reset_index(drop=False)
hazard_2020_count
```

```python
# hazard_2020_count = hazard_2020_count.groupby('사업자번호', as_index=False).count()
hazard_2020_count.rename(columns = {'등록일시':'건수'}, inplace=True)
hazard_2020_count
```

```python
hazard_2020_merged = pd.merge(kosdaq_df, hazard_2020_count, how='left', on='사업자번호')
# hazard_2020_merged.rename(columns = {'등록일시':'건수'}, inplace=True)
hazard_2020_merged
```
