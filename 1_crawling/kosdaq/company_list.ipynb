```python
import FinanceDataReader
import pandas as pd
```

```python
def get_kosdaqlist(start_date='2019-01-01', end_date='2021-12-31'):
    kosdaq = FinanceDataReader.StockListing("KOSDAQ")
    kosdaq_list =  kosdaq[(kosdaq['ListingDate'] <= pd.to_datetime(end_date)) & (kosdaq['ListingDate'] >= pd.to_datetime(start_date))]
    return kosdaq_list
```
