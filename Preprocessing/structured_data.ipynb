## 1. 라이브러리 및 기업리스트 데이터 불러오기

```python
import pandas as pd
import numpy as np
```

```python
green_2020 = pd.read_csv("한국환경산업기술원_녹색제품인증기업현황_20201001.csv", encoding='cp949')
green_2021 = pd.read_excel("green2021.xlsx")
kosdaq_list = pd.read_excel("C:/Users/hanah/Desktop/workspace/Project/ESG/crawling/dart/kosdaq_report_final.xlsx", index_col=0)
```

## 2. 20년, 21년도 데이터 병합

```python
total_green = pd.merge(green_2021, green_2020, left_on = green_2021.업체명, right_on = green_2020.업체명, how='left')
total_green.drop(['key_0', '업체명_y'], axis=1, inplace=True)
total_green
```

```python
total_green.info()
```

```python
#21년도 인증기업은 주소데이터가 없으므로 해당 컬럼을 기준으로 년도 구분
total_green['주소'].loc[total_green['주소'].isnull() == False] = 2020 
total_green['주소'].loc[total_green['주소'].isnull() == True] = 2021
total_green['주소'].value_counts()
```

```python
total_green.drop(['전화번호', '기업주소', '상세주소', '대표전화', '팩스', '홈페이지'], axis=1, inplace=True)
total_green.rename(columns={'업체명_x':'회사명'}, inplace=True)
total_green.rename(columns={'주소':'녹색제품인증연도'}, inplace=True)
total_green
```

```python
total_green.to_excel('total_green.xlsx')
```

## 3. 코스닥 기업리스트에 병합

```python
kosdaq_list = pd.merge(kosdaq_list, total_green[['회사명', '녹색제품인증연도']], how='left')
kosdaq_list['녹색제품인증연도'].value_counts()
```

```python
kosdaq_list.to_excel('kosdaq_report_green.xlsx')
```
