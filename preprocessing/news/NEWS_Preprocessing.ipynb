### 통합 뉴스 데이터 불러오기

```python
import re
import numpy as np
import pandas as pd
import string

df = pd.read_csv('total_news.csv')

df.shape
```

### 본문 데이터 비어 있는 행 제거

```python
df = df[~df['Content'].isnull()]
df.shape
```

### 인사동정 기사 제거

```python
# 인사동정 기사 제거
df = df[
        ~df['Content'].str.contains('오늘의 인사') 
        & ~df['Content'].str.contains('\[인사\]')
        & ~df['Content'].str.contains('일 인사')
        & ~df['Content'].str.contains('\[부고\]')
        ]
df.shape
```

### 영어기사 제외(140자이상 영어)

```python
def text_filter(text):
    import re
    from string import whitespace, punctuation
    punct_except_percent = ''.join([chr for chr in punctuation if chr != '%'])
    whitespace_convert_pattern = re.compile(f'[{whitespace}{punct_except_percent}]+')
    exclude_pattern = re.compile(r'^[a-zA-Z]*$/;')

    # exclusions = exclude_pattern.findall(text)
    # print(exclusions)
    result = whitespace_convert_pattern.sub(' ', text)
    result = exclude_pattern.sub(' ', result).strip()
    result = whitespace_convert_pattern.sub(' ', result)
    return result

# df['text'] = df[df['Content'].map(text_filter).str.len() > 140]
# df['text'].shape

df = df[df['Content'].map(text_filter).str.len() > 140]
df.shape
```

### 뉴스 Content 기타 전처리 단어, 공백 함수화 제거

```python
def preprocessing(text):
    repl =''
    
    
    reporter_pattern = re.compile(r"([가-힣]{2,5} 기자)|([가-힣]{2,5}기자)")
    text = reporter_pattern.sub('', text)

    pattern = '([ㄱ-ㅎㅏ-ㅣ]+)' # 자음, 모음 제거
    text = re.sub(pattern= pattern, repl=repl, string=text)

    pattern = '[^가-히\s]' # 특수기호 제거
    text = re.sub(pattern= pattern, repl=repl, string=text)

    pattern = '<[^>]*>' # html 제거
    text = re.sub(pattern = pattern, repl=repl,string=text)

    pattern = '\.([^\.]*(?:기자|뉴스|특파원|교수|작가|대표|논설|주필|에세이|화백|사설|소장|단장|큐레이터|저작권|평론가|©|©|ⓒ|\@|\/|=|:앞쪽_화살표:|무단|전재|재배포|금지|\[|\]|\(\))[^\.]*)$'
    text = re.sub(pattern = pattern, repl=repl,string=text) # 바이라인 제거


    symbols = string.punctuation.replace(".", "").replace("?", "").replace("!", "") + "·ㆍ■◆△▷▶▼�""''…※↑↓▲☞ⓒ⅔"
    text = text.translate(str.maketrans("", "", symbols))


    text = re.sub('\d+\.\d*', '',text) #숫자, 소숫점 제거

    text = re.sub("([a-zA-Z0-9-]+(\@|\.)[a-zA-Z0-9-.]+)", '', text) #이메일, URl 제거
    
    text = re.sub("\.+\.", '.', text) #다수의 점 (ex : ...) 점 한개로 대체
    
    text = re.sub(' +', ' ', text) #다수의 공백 축소

    text = text.replace('Copyrights', '').replace('무단 전재 및 재배포 금지', '')

    return text

df['Content'] = df['Content'].map(preprocessing)
df.shape
df
```

```python
df.to_csv('NEWS_Preprocessing2.csv', index=False)
```
